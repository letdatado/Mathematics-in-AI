# 3.2 Dot Product

```python
import numpy as np

a = np.array([3, 4])
b = np.array([4, 3])
dot = np.dot(a, b)
```

```python
print(f"Vector a: {a}")
print(f"Vector b: {b}")
print(f"Dot product a●b: {dot}")
```

# 3.5 From Dot Product to Angle

```python
theta = np.arccos(np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b)))
```

```python
print(f"Angle between a and b (degrees): {np.degrees(theta):.2f}")
```

# 3.7 Orthogonality — When Two Vectors Have Nothing in Common

```python
a = np.array([1, 0])
b = np.array([0, 1])
```

```python
print(f"Vector a: {a}")
print(f"Vector b: {b}")
print(f"Dot product a●b: {np.dot(a,b)}")
```

# 3.8 Cross Product — “The Perpendicular Vector”

```python
a = np.array([2, 3, 4])
b = np.array([5, 6, 7])
cross = np.cross(a, b)
```

```python
print(f"Vector a: {a}")
print(f"Vector b: {b}")
print(f"Cross product axb: {cross}")
```
