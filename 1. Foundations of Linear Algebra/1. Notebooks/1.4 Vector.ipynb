```python
import numpy as np
import plotly.graph_objects as go

v = np.array([3, 4])

fig = go.Figure()

# Add the vector as an arrow
fig.add_trace(go.Scatter(
    x=[0, v[0]],
    y=[0, v[1]],
    mode="lines",
    line=dict(color="red", width=4),
    marker=dict(size=6, color="red"),
    name="v"
))


fig.update_layout(
    title="Vector v = [3, 4]",
    xaxis=dict(range=[-1, 5], zeroline=True),
    yaxis=dict(range=[-1, 5], zeroline=True, scaleanchor="x", scaleratio=1),
    width=500,
    height=500,
    template="plotly_white"
)

# Arrow for v (from (0,0) to (3,4))
fig.add_annotation(
    x=v[0], y=v[1],        # arrow head (data coords)
    ax=0, ay=0,            # arrow tail (data coords)
    xref="x", yref="y",
    axref="x", ayref="y",
    showarrow=True,
    arrowhead=3,
    arrowsize=1,
    arrowwidth=2,
    arrowcolor="red",
    text=""                # no text, just arrow
)

 
fig.show()
```
