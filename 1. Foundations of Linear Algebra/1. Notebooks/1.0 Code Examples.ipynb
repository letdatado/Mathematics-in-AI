# 1.7 Distance Between Two Points

```python
p = np.array([1, 2, 3])
q = np.array([4, 6, 3])

euclidean = np.linalg.norm(p - q)             # L2
manhattan = np.linalg.norm(p - q, ord=1)      # L1
chebyshev = np.linalg.norm(p - q, ord=np.inf) # L∞

```

```python
print(f"Point p: {p}")
print(f"Point q: {q}")
print("-"*25)
print(f"Euclidean Distance L2: {euclidean}")
print(f"Manhattan Distance L1: {manhattan}")
print(f"Chebyshev Distance L∞: {chebyshev}")
```
