# 6.5 The workhorse equations (with shapes)
## 6.5A Linear regression (batch prediction)

```python
import numpy as np

# Example pattern (yours may differ!)
rng = np.random.default_rng(0)
mean = np.array([1.0, -1.0])
cov = np.array([[1.5, 0.8],
                [0.8, 1.0]])
X = rng.multivariate_normal(mean, cov, size=120)
```

```python
w = np.array([0.8, -0.3])   # n=3
b = 0.2
y_hat = X @ w + b 
```

```python
print(f"w: {w}")
print(f"b: {b}")
print(f"Y Pred: {y_hat}")
```

## 6.5B Cosine Similarity (Rows normalized)

```python
X_norm = X / np.linalg.norm(X, axis=1, keepdims=True)  # (m,n)
S = X_norm @ X_norm.T                                  # (m,m) similarities
S
```

# 6.6 Centering & scaling
## 6.6A Centering (remove mean per feature)

```python
mu = X.mean(axis=0, keepdims=True)  # (1,n)
Xc = X - mu
print(Xc)
```

## 6.6B Standardizing (unit variance per feature)

```python
sigma = X.std(axis=0, keepdims=True) + 1e-8
Z = (X - mu) / sigma
print(Z)
```

# 6.8 Tiny end-to-end demo (fit a line/plane just to see shapes)

```python
# closed-form w for least squares on 2 features (no reg), centered X
# y target is synthetic
y_target = 0.7*X[:,0] - 0.2*X[:,1] + rng.normal(0,0.3, size=X.shape[0])

# Normal equation: w = (X^T X)^{-1} X^T y  (add small ridge for stability)
lam = 1e-6
XT_X = X.T @ X + lam*np.eye(X.shape[1])
w_hat = np.linalg.solve(XT_X, X.T @ y_target)
b_hat = 0.0  # skip intercept here; or add a bias column of 1s

print("w_hat:", w_hat)
```
