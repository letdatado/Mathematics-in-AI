# 8.4 Variance Explained

```python
import numpy as np
from numpy.linalg import eig

# synthetic correlated 2D data
rng = np.random.default_rng(1)
X = rng.multivariate_normal(mean=[0,0],
                            cov=[[3, 2.5],
                                 [2.5, 3]], size=150)
# center it
Xc = X - X.mean(axis=0, keepdims=True)

# covariance & eigendecomposition
Sigma = (Xc.T @ Xc) / (Xc.shape[0]-1)
eigvals, eigvecs = eig(Sigma)

# sort by descending eigenvalue
idx = np.argsort(eigvals)[::-1]
eigvals, eigvecs = eigvals[idx], eigvecs[:, idx]

# principal components
pc1, pc2 = eigvecs[:,0], eigvecs[:,1]
explained = eigvals / np.sum(eigvals)

print(f"Variance Explained PC1: {explained[0]}\nVariance Explained PC2: {explained[1]}")
```
