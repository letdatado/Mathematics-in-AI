```python
import numpy as np
import plotly.graph_objects as go

# -----------------------------------------
# Data
# -----------------------------------------
x = np.array([1, 2, 3])
y = np.array([2, 2.9, 4.2]).reshape(-1, 1)

# Design matrix
X = np.vstack([np.ones_like(x), x]).T

# Normal equation
w = np.linalg.inv(X.T @ X) @ X.T @ y
w0, w1 = w.ravel()
print("w0 (bias), w1 (slope) =", w0, w1)

# Predictions
y_hat = X @ w
y_hat = y_hat.flatten()

# -----------------------------------------
# Plotly figure
# -----------------------------------------
fig = go.Figure()

# Data points
fig.add_trace(
    go.Scatter(
        x=x,
        y=y.flatten(),
        mode="markers",
        name="Data",
        marker=dict(size=10, color="blue"),
        hovertemplate="x=%{x}<br>y=%{y}<extra></extra>",
    )
)

# Best fit line
fig.add_trace(
    go.Scatter(
        x=x,
        y=y_hat,
        mode="lines",
        name="Best Fit Line",
        line=dict(color="red", width=3),
        hovertemplate="x=%{x}<br>ŷ=%{y:.3f}<extra></extra>",
    )
)

# -----------------------------------------
# Layout
# -----------------------------------------
fig.update_layout(
    title=(
        "Linear Regression as Orthogonal Projection<br>"
        f"<span style='font-size:12px;color:gray;'>"
        f"w₀ = {w0:.3f},  w₁ = {w1:.3f}"
        "</span>"
    ),
    xaxis_title="x",
    yaxis_title="y",
    width=700,
    height=500,
    template="plotly_white",
    legend=dict(yanchor="top", y=0.98, xanchor="left", x=0.02),
)

fig.show()
```
